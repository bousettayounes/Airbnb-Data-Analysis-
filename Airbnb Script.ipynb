# Data Preprocessing

```python
Data.columns=[col.lower().replace(' ','_') for col in Data.columns]
```

```python
Data.columns
```

```python
Missing_value_perc= (Data.isnull().sum()/Data.shape[0])*100
Missing_value_perc
```

```python
Missing_value=pd.DataFrame({'column':Missing_value_perc.keys(),'Missing Value %':Missing_value_perc})
Missing_value.reset_index(drop=True,inplace=True)
Missing_value
```

```python
Data["id"].unique().shape[0]/Data["id"].shape[0]*100
```

```python
Data["host_id"].unique().shape[0]/Data["host_id"].shape[0]*100
```

Most of these values ​​are distinct for each line. I don't think it makes sense for analysis, so I'll delete them.

```python
Data.drop(columns=["id","host_id"],axis=1,inplace=True)
```

```python
Data["country_code"].value_counts()
```

```python
Data["country"].value_counts()
```

```python
Data.drop(columns=["country_code","country"],axis=1,inplace=True)
```

```python
def remove_dollar_sign(value):
    if pd.isna(value):
        return np.NaN
    else:
        return float(value.replace("$","").replace(",","").replace(" ",""))
```

```python
Data["price"]
```

```python
Data["price"]=Data["price"].apply(lambda x: remove_dollar_sign(x))
```

```python
Data["price"]
```

```python
Data["service_fee"]=Data["service_fee"].apply(lambda x: remove_dollar_sign(x))
Data["service_fee"]
```

```python
Data["last_review"]=pd.to_datetime(Data["last_review"])
```

```python
Data.drop(columns=["house_rules","license"], axis=1, inplace=True)
```
